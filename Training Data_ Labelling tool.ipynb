```python
import numpy as np
import cv2
import tkinter as tk
from PIL import Image, ImageTk
import glob
import os
import copy
```

```python
animals=['giraffe','kangaroo','rabbit','cow','sheep','deer','rhinoceros','seal','goat','donkey','dolphine']
image_names=glob.glob(f'./data_image/{animal}/*.jpg')
descriptions_dict={}
with open(f'./data_image/{animal}/{animal}.txt','r') as file:
    while True:
        line=file.readline()
        if line=='':
            break
        file_path,description=line.split(':')
        descriptions_dict[file_path]=description[:-1]
using_words=[]
with open('./사용할 단어모음.txt') as f:
    while True:
        line=f.readline()[:-1]
        if line=='':
            break
        using_words.append(line)
while '---' in using_words:
    del using_words[using_words.index('---')]
```

```python
class app():
    def __init__(self,animal):
        image_names=glob.glob(f'./data_image/{animal}/*.jpg')
        descriptions_dict={}
        with open(f'./data_image/{animal}/{animal}.txt','r') as file:
            while True:
                line=file.readline()
                if line=='':
                    break
                file_path,description=line.split(':')
                descriptions_dict[file_path]=description[:-1]
        using_words=[]
        with open('./사용할 단어모음.txt') as f:
            while True:
                line=f.readline()[:-1]
                if line=='':
                    break
                using_words.append(line)
        while '---' in using_words:
            del using_words[using_words.index('---')]
        self.word_list=copy.copy(using_words)
        self.descriptions_dict=copy.copy(descriptions_dict)
        self.image_paths=copy.copy(image_names)
        self.right_description=[]
# number of saved training image
        self.n=len(glob.glob(f'./final_train_image/{animal}/*.jpg'))
        self.cur_image=None
        self.root=None
        self.run()
    def save_button(self):
        delete_path=self.image_paths[0]
        delete_img=self.image_paths[0].split('\\')[1]
        os.remove(delete_path)
        del self.descriptions_dict[delete_img]
        _=self.image_paths.pop(0)
        self.n+=1
        new_path=f'./final_train_image/{animal}/{animal}_{self.n}.jpg'
        right_description=self.input_description.get()
        self.right_description.append(new_path+':'+right_description)
        cv2.imwrite(new_path,self.cur_image)
        self.root.destroy()
        self.run()
    def delete_button(self):
        delete_path=self.image_paths[0]
        delete_img=self.image_paths[0].split('\\')[1]
        os.remove(delete_path)
        del self.descriptions_dict[delete_img]
        _=self.image_paths.pop(0)
        self.root.destroy()
        self.run()
    def add_word_button(self):
        new_word=self.input_word.get()
        if new_word!='' and new_word not in self.word_list:
            self.word_list.append(new_word)
            self.root.destroy()
            self.run()
        elif new_word in self.word_list:
            self.word_list.remove(new_word)
            self.root.destroy()
            self.run()
    def quit_button(self):
        self.root.destroy()
        with open(f'./data_image/{animal}/{animal}.txt','w') as f:
            for file_name,description in self.descriptions_dict.items():
                f.write(file_name+':'+description+'\n')
        with open(f'./final_train_image/{animal}/{animal}.txt','a') as f:
            for description in self.right_description:
                f.write(description+'\n')
        with open(f'./사용할 단어모음.txt','w') as f:
            for word in self.word_list:
                f.write(word+'\n')
    def run(self):
        self.root=tk.Tk()
        if len(self.descriptions_dict)==0:
            self.quit_button()
        self.button_save=tk.Button(self.root,text='Save',command=self.save_button,padx=150,pady=10)
        self.button_delete=tk.Button(self.root,text='Delete',command=self.delete_button,padx=50,pady=10)
        self.button_add_words=tk.Button(self.root,text='A_or_D_Word',command=self.add_word_button,pady=10)
        self.button_quit=tk.Button(self.root,text='Quit',command=self.quit_button,padx=50,pady=10)
        self.input_word=tk.Entry(self.root,width=10)
        self.description_text=tk.Label(self.root,text=self.descriptions_dict[self.image_paths[0].split('\\')[1]])
        self.input_description=tk.Entry(self.root,width=50)
        self.words_text=tk.Label(self.root,text='\n'.join(self.word_list))

        self.cur_image = cv2.imread(self.image_paths[0])
        self.cur_image=cv2.resize(self.cur_image,(300,300),interpolation=cv2.INTER_AREA)
        b,g,r = cv2.split(self.cur_image)
        img = cv2.merge((r,g,b))
        img = Image.fromarray(img)
        imgtk = ImageTk.PhotoImage(image=img)
        self.image_label=tk.Label(self.root, image=imgtk)

        self.words_text.grid(row=0,column=0)
        self.image_label.grid(row=0,column=1)
        self.description_text.grid(row=0,column=2)
        self.input_description.grid(row=2,column=2)
        self.input_word.grid(row=2,column=0)
        self.button_delete.grid(row=3,column=1)
        self.button_add_words.grid(row=3,column=0)
        self.button_save.grid(row=3,column=2)
        self.button_quit.grid(row=0,column=3)
        self.root.title('Captioning Labelling tool')
        self.root.attributes('-fullscreen',True)
        self.root.mainloop()
animal=input('Enter animal to label: ')
app=app(animal)
```
